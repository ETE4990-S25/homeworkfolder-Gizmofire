````python
class Animal:
  def __init__(self, name, age, spe, edi):
    self.name = name
    self.age = age
    self.species = spe
    self.edible = edi

class Food(Animal):
  def __init__(self, name, age, spe, owner):
    super().__init__(name, age, spe, "True")
    self.ownerName = owner
    self.type = "food"

class Pet(Animal):
  def __init__(self, name, age, spe, owner):
    super().__init__(name, age, spe, "Pleas dont")
    self.ownerName = owner
    self.type = "pet"

class Wild(Animal):
  def __init__(self, name, age, spe):
    super().__init__(name, age, spe, "maybe")
    self.ownerName = "wilderness"
    self.type = "wild"

listOfAnimals = []

listOfAnimals.append(Pet("Fido", 5, "Dog", "Bob"))
listOfAnimals.append(Food("Beef", 1, "Cow", "Bob"))
listOfAnimals.append(Wild("Wolfie", 2, "Wolf"))
listOfAnimals.append(Pet("Fluffy", 3, "Cat", "Alice"))
listOfAnimals.append(Food("Chick", 1, "Chicken", "Alice"))

# for animal in listOfAnimals:
#     print("Name: " + animal.name)
#     print("Age: " + str(animal.age))
#     print("Species: " + animal.species)
#     print("Edible: " + animal.edible)
#     print("Owner: " + animal.ownerName)
#     print("Type: " + animal.type)
#     print("\n\n")


def main():
  exit = True

  while (exit):
    print("1. Add a new animal")
    print("2. List all animals")
    print("3. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
      name = input("Enter the name of the animal: ")
      age = input("Enter the age of the animal: ")
      species = input("Enter the species of the animal: ")
      owner = input("Enter the owner of the animal: ")
      type = input("Enter the type of the animal. \n Options: (1) pet animal (2) food animal (3) wild animal: ")

      if type == "pet":
        listOfAnimals.append(Pet(name, age, species, owner))
      elif type == "food":
        listOfAnimals.append(Food(name, age, species, owner))
      elif type == "wild":
        listOfAnimals.append(Wild(name, age, species))
      else:
        print("Invalid type")

    elif choice == "2":
      for animal in listOfAnimals:
        print("Name: " + animal.name)
        print("Age: " + str(animal.age))
        print("Species: " + animal.species)
        print("Edible: " + animal.edible)
        print("Owner: " + animal.ownerName)
        print("Type: " + animal.type)
        print("\n\n")

    elif choice == "3":
      exit = False

    else:
      print("Invalid choice")



main()




  

````
